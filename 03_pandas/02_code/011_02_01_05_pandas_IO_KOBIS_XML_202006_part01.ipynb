```python
# 위와 같이 한다면, 정보들이 완벽할 때에만 이렇게 가능하다. 아닌 경우에는 정보들이 밀릴 수 있다!!!
# --> 수정된 코드!!!!!
# ID 리스트에 대한 정보를 어떻게 얻을 지 확인하였으면, 이에 대한 작업 부분
movieIDDF = pd.DataFrame()
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append({"movieCd":"", "movieNm" : "", "movieNmEn" : "", "openDt":"","directorpeopleNm":""},ignore_index=True)


num = len(soup.find_all("movie"))
for i in range(0, num):
    movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("movieCd")] = soup.find_all("movie")[i].find("moviecd").text
    movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("movieNm")] = soup.find_all("movie")[i].find("movienm").text
    movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("movieNmEn")] = soup.find_all("movie")[i].find("movienmen").text 
    movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("openDt")] = soup.find_all("movie")[i].find("opendt").text
    if soup.find_all("movie")[i].find_all("peoplenm") != [] :
        movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("directorpeopleNm")] = soup.find_all("movie")[i].find_all("peoplenm")[0].text
    else:
        movieIDDF_temp.iat[0,movieIDDF_temp.columns.get_loc("directorpeopleNm")] = ""
        
    # 합치기
    movieIDDF = movieIDDF.append(movieIDDF_temp,ignore_index=True)
    
    # 다시 초기화
    for j in range(0, len(movieIDDF.columns)):
        movieIDDF_temp.iat[0,j]=""
movieIDDF
```

```python
# 앞에서와 같은 추가 정보 생성
movieIDDF["year"] = movieIDDF["openDt"].apply(lambda x: str(x)[:4])
movieIDDF["month"] = movieIDDF["openDt"].apply(lambda x: str(x)[4:6])
movieIDDF["day"] = movieIDDF["openDt"].apply(lambda x: str(x)[6:])

movieIDDF = movieIDDF.sort_values(by ="openDt")
movieIDDF = movieIDDF.set_index("movieCd")
movieIDDF = movieIDDF[["openDt","year","month","day", "movieNm","movieNmEn","directorpeopleNm"]]
movieIDDF
```

```python
# 상세 정보 조회 및 요청해서 원하는 파일 생성
movieIDDF["type"] = ""
movieIDDF["ActorNum"] = ""
movieIDDF["cast"] = ""
movieIDDF["cast_peopleNm"] = ""
movieIDDF["genre"] = ""

for i in movieIDDF.index:
    tempURL_Part1 = """http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?key="""
    keyNum = """430156241533f1d058c603178cc3ca0e"""
    tempURL_Part2 = """&movieCd="""
    movieId = i
    
    tempURL = tempURL_Part1 + keyNum + tempURL_Part2 + movieId
    
    # For XML Type 
    res = urllib.request.urlopen(tempURL)
    tempsoup = BeautifulSoup(res, "html.parser")

    if tempsoup.find_all("typenm")!=[]:
        movieIDDF.at[i,"type"] = tempsoup.find_all("typenm")[0].text
    if tempsoup.find_all("actor")!=[]:
        movieIDDF.at[i,"ActorNum"] = len(tempsoup.find_all("actor"))
    if tempsoup.find_all("actor")!=[]:
        movieIDDF.at[i,"cast"] = tempsoup.find_all("actor")[0].find("cast").text
        movieIDDF.at[i,"cast_peopleNm"] =tempsoup.find_all("actor")[0].find("peoplenm").text
    if tempsoup.find("genres").text!="":
        movieIDDF.at[i,"genre"] = tempsoup.find("genres").find("genre").find("genrenm").text
movieIDDF
```
