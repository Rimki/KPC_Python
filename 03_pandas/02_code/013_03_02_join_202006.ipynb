# Join

```python
import pandas as pd
import numpy as np
```

```python
#01) DataSet
baseball_player = pd.DataFrame(
    {
        "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993, 2001, 2001,2010, None]
    }
)
baseball_team = pd.DataFrame(
    {
         "team_Name":["한화","두산","기아","넥센"],
        "team_home" :["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    }
)
```

```python
baseball_player
```

```python
baseball_team
```

```python
# Left Outter Join 
pd.merge(baseball_player, baseball_team, how='left', 
         left_on = "team_No", right_on="team_id")
```

```python
# Inner Join 
pd.merge(baseball_player, baseball_team, how='inner', 
         left_on = "team_No", right_on="team_id")
```

```python
# right outter
pd.merge(baseball_player, baseball_team, 
         how='right', left_on = "team_No", 
         right_on="team_id")
```

```python
# left excluding join
# 01) left join 
temp = pd.merge(baseball_player, baseball_team, 
                how='left', 
                left_on = "team_No", 
                right_on="team_id")
# 02) select NaN
temp[temp["team_id"].isnull()]
```

```python
# right excluding join
#01) right join
temp = pd.merge(baseball_player, baseball_team, 
                how='right', 
                left_on = "team_No", 
                right_on="team_id")
# 02) select NaN
temp[temp["player_Name"].isnull()]
```

```python
# union
pd.merge(baseball_player, baseball_team,
         how='outer',
         left_on = "team_No", 
         right_on="team_id")
```

```python
# full outer excluding
temp = pd.merge(baseball_player, baseball_team,
                how='outer', 
                left_on = "team_No", 
                right_on="team_id")
temp[temp["team_No"].isnull() | 
     temp["team_id"].isnull()]
```

```python
baseball_player = pd.DataFrame(
    {
        "player_Name":["권혁","이범호","필","박병호","양준혁"],
        "team_No":[1993, 2001, 2001,2010, None]
    },
    index = ["p_01","p_02","p_03","p_04","p_05"]
)
baseball_player.index.name="player_id"

baseball_team = pd.DataFrame(
    {
         "team_Name":["한화","두산","기아","넥센"],
        "team_home" :["대전","서울","광주","고척"]
    },
        index = [1993,1999,2001,2010]
)
```

```python
baseball_player
```

```python
baseball_team
```

```python
# Left Outter Join 
pd.merge(baseball_player, baseball_team, 
         how='left',
         left_on = "team_No",
         right_index=True)
```

```python
# Inner Join 
pd.merge(baseball_player, baseball_team, how='inner', 
         left_on = "team_No", right_index=True)
```

```python
# right outter
pd.merge(baseball_player, baseball_team, how='right', 
         left_on = "team_No", right_index=True)
```

```python
# right outter
pd.merge(baseball_player, baseball_team, how='outer', 
         left_on = "team_No", right_index=True)
```

```python
temp = pd.merge(baseball_player, baseball_team, how='outer', left_on = "team_No", right_index=True)
temp
```

```python
temp[temp.loc[:,"player_Name"].isnull()  | temp.loc[:,"team_Name"].isnull()]
```
